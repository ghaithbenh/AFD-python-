```python
#les continent mayet3awdoch
#regroupement par continent
df3=df1.groupby('continent').mean()
df3
```

```python
#library pour la représentation graphiques
from matplotlib import pyplot as plt

plt.plot(df3['continent'],df3['total_deaths'])
plt.show()
#impossible de faire la representaion graphique a cause decalage 
#3awed el indexation mta3 dataframe 
```

```python
df4=df1.groupby('continent').mean().reset_index()
#reset indices to match new data frame
#to reformat the data frame indicies 
#bech ywali mafamech decalage 
df4
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(df4['continent'],df4['total_deaths'])
plt.show()
```

```python
#création de graphique avec la méthode plot 
plt.plot(df4['continent'],df4['total_deaths'])
plt.xlabel("continent")
plt.ylabel("total deaths")
plt.title("means deaths per continent")
#affichage de graphique 
plt.show()
```
