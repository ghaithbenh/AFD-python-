 <h2> <u>Exercice 2 </u> </h2>

```python
x=np.linspace(0,10,100) #6valeurs #100valeurs #30valeurs
#creation du tracé d la fonction f(x) :
f=3*x+5*np.cos(3*x)
plt.plot(x,f,label="f(x)")
plt.xlabel("Les x")
plt.ylabel("f(x)")
plt.legend(["f(x)"])
plt.title("La fonction f(x) ")
#affichage de graphique 
plt.show()
```

```python
# creation du tracé d la fonction g(x) :
x=np.linspace(0,10,500)
g=np.tan(x+2)
plt.plot(x,g)
plt.xlabel("Les x")
plt.ylabel("g(x)")
plt.legend(["g(x)"])
plt.title("La fonction g(x) ")
#affichage de graphique 
plt.show()
```

```python
x=np.linspace(0,10,500)
h=np.exp(3*x+3)
plt.plot(x,h,"r--",label="z(x) ")
plt.xlabel("Les x")
plt.ylabel("z(x)")
plt.legend()
plt.title("La fonction z(x) ")
#affichage de graphique 
plt.show()
```
